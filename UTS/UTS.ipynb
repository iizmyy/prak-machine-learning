```python
df_Salary = pd.read_csv('Salary.csv')
X = df_Salary.loc[:, ('Age','Education Level','Years of Experience','Senior')]
y = df_Salary.loc[:, ('Salary')]


# Membuat objek regresi linier
regr_b = linear_model.LinearRegression()

# Training model dengan data latih
regr_b.fit(X.values, y)

# Nilai koefisien dan nilai intercept
print("Koefisien : ", regr_b.coef_)
print("Intercept : ", regr_b.intercept_)

# Prediksi data
data_uji = np.array([[20.0,1,1.0,0]])
pred = regr_b.predict(data_uji)
print("\nPrediksi : ",pred)
```
