```python
import numpy as np
import pandas as pd
from sklearn import linear_model

# Load dataset CAR DETAILS FROM CAR DEKHO
df_mobil = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
```

```python
# Training Data
mobil_X_train = df_mobil.loc[:200, ('year','km_driven','seller_type')]
mobil_X_train = pd.get_dummies(mobil_X_train, prefix='ST',columns=['seller_type'])
mobil_y_train = df_mobil.loc[:200,'selling_price']
```

```python
# Membuat objek regresi linier
regr_b = linear_model.LinearRegression()

# Training model dengan data latih
regr_b.fit(mobil_X_train.values, mobil_y_train)

# Nilai koefisien dan nilai intercept
print("Koefisien : ", regr_b.coef_)
print("Intercept : ", regr_b.intercept_)
```

```python
# Prediksi harga sebuah mobil dengan year:2015, km_driven:50000, seller_type : Individual
mobil_x_test1 = np.array([[2015,50000,1,0]])
harga_pred_mobil = regr_b.predict(mobil_x_test1)
print("\nPrediksi mobil year=2015, km_driven=50000, seller_type=Individual :", harga_pred_mobil, "M")

# Prediksi harga beberapa mobil
mobil_X_test2 = df_mobil.loc[101:155, ('year','km_driven','seller_type')]
mobil_X_test2 = pd.get_dummies(mobil_X_test2, prefix='ST',columns=['seller_type'])
harga_pred_mobil = regr_b.predict(mobil_X_test2.values)

# Hasil Prediksi dalam tabel
print("\nHasil Prediksi data test 2")
mobil_Y_test2 = df_mobil.loc[101:155, 'seller_type']
mobil_test2 = pd.concat([mobil_X_test2, mobil_Y_test2], axis=1)
mobil_test2['HARGA_PRED'] = harga_pred_mobil
print(mobil_test2)
```
