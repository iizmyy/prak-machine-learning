```python
df_lifespan = pd.read_csv('Life_Expectancy_Data.csv')
display(df_lifespan.head())
df_lifespan.shape
```

```python
# Training Data
lifespan_x_train = df_lifespan.loc[:, ('Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles ',' BMI ','under-five deaths ','Polio','Total expenditure','Diphtheria ',' HIV/AIDS','GDP','Population',' thinness  1-19 years',' thinness 5-9 years','Income composition of resources','Schooling')]
lifespan_y_train = df_lifespan.loc[:, 'Life expectancy ']

# Membuat objek regresi linier
regr_b = linear_model.LinearRegression()

# Training model dengan data latih
regr_b.fit(lifespan_x_train.values, lifespan_y_train)

# Nilai koefisien dan nilai intercept
# print("Koefisien : ", regr_b.coef_)
# print("Intercept : ", regr_b.intercept_)

# Prediksi life expectancy Indonesia
indonesia_data = np.array([[179,119,0.09,200.0684443,78,12943,26.5,142,8,2.85,78,0.3,3491.595887,2.55131116e8,1.4,1.2,0.682,12.9]])
indonesia_lifespan_pred = regr_b.predict(indonesia_data)
print("\nPrediksi life expectancy Indonesia :", indonesia_lifespan_pred)

# Prediksi life expectancy Jerman
jerman_data = np.array([[69,2,11.03,941.7562909,88,443,61.9,3,94,11.3,95,0.1,4792.65288,89825.0,1.1,1.1,0.92,17.0]])
jerman_lifespan_pred = regr_b.predict(jerman_data)
print("\nPrediksi life expectancy Jerman :", jerman_lifespan_pred)

```
