```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load dataset COVID-19 Indonesia
from google.colab import files
uploaded = files.upload()
df_covid = pd.read_csv('covid_19_indonesia_time_series_all.csv')
```

```python
# Menampilkan sebagian dari tabel (5 baris pertama)
df_covid_head = df_covid.head()
print(df_covid_head)
```

```python
# Memberikan nama kolom pada DataFrame
df_covid.columns = ['Date', 'Location ISO Code', 'Location', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 'Total Recovered', 'Total Active Cases', 'Location Level', 'City or Regency', 'Province', 'Country', 'Continent', 'Island', 'Time Zone', 'Special Status', 'Total Regencies', 'Total Cities', 'Total Districts', 'Total Urban Villages', 'Total Rural Villages', 'Area (km2)', 'Population', 'Population Density', 'Longitude', 'Latitude', 'New Cases per Million', 'Total Cases per Million', 'New Deaths per Million', 'Total Deaths per Million', 'Total Deaths per 100rb', 'Case Fatality Rate', 'Case Recovered Rate', 'Growth Factor of New Cases', 'Growth Factor of New Deaths']
# Pilih kolom yang akan digunakan untuk prediksi (misalnya: New_Cases, Total_Cases, dll.)
feature_cols = ['New Cases']  # Ganti dengan kolom yang sesuai

# Variabel independen (X) dan dependen (y)
X = df_covid[feature_cols]
y_recovery = df_covid['New Recovered']
y_deaths = df_covid['New Deaths']

# Inisialisasi model regresi linear dengan fitur nama
model_recovery = LinearRegression(fit_intercept=True)
model_deaths = LinearRegression(fit_intercept=True)

# Melatih model dengan seluruh data
model_recovery.fit(X, y_recovery)
model_deaths.fit(X, y_deaths)

# Nilai koefisien dan nilai intercept
print("Koefisien Recovery:", model_recovery.coef_)
print("Intercept Recovery:", model_recovery.intercept_)

print("\nKoefisien Deaths:", model_deaths.coef_)
print("Intercept Deaths:", model_deaths.intercept_)
```

```python
# Prediksi menggunakan data baru
new_data = np.array([[100]])  # Ganti dengan nilai yang sesuai
recovery_pred = model_recovery.predict(new_data)
deaths_pred = model_deaths.predict(new_data)

print("\nPrediksi Jumlah Recovery untuk Kasus Baru:", recovery_pred[0])
print("Prediksi Jumlah Kematian untuk Kasus Baru:", deaths_pred[0])
```

```python
# Membuat DataFrame untuk menampilkan hasil prediksi
result_df = pd.DataFrame({
    'New Cases': new_data.flatten(),
    'Predicted Recovery': recovery_pred.flatten(),
    'Predicted Deaths': deaths_pred.flatten()
})

print(result_df)
```
